```python
import torch
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from transformers import BertTokenizer, BertModel
```

```python
# Load the dataset
dataset_path = "/content/Final_Augmented_dataset_Diseases_and_Symptoms.csv"
df = pd.read_csv(dataset_path)

# Display first few rows
print(df.head())
```

```python
# Load BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased", output_attentions=True)
```

```python
def get_attention_scores(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)

    # Get model outputs with attention weights
    with torch.no_grad():
        outputs = model(**inputs)

    attention = outputs.attentions  # Extract attention weights
    attention = torch.stack(attention).squeeze(1).cpu().numpy()  # Convert to numpy
    return attention, inputs
```

```python
def plot_attention_heatmap(text):
    attention, inputs = get_attention_scores(text)

    tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"].squeeze().tolist())

    # Taking attention from the last layer
    attention_matrix = attention[-1].mean(axis=0)

    # Convert to a heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(attention_matrix, xticklabels=tokens, yticklabels=tokens, cmap="coolwarm", annot=True)
    plt.title("BERT Attention Heatmap")
    plt.show()
```

```python
# Example 1: Severe Symptoms
text1 = "Patient reports persistent cough, high fever, and difficulty breathing for the past three days."
plot_attention_heatmap(text1)

# Example 2: Mild Symptoms
text2 = "Mild headache and occasional dizziness, but no fever or cough."
plot_attention_heatmap(text2)

# Example 3: Rare Important Symptoms
text3 = "The patient exhibits cyanosis, extreme fatigue, and rapid heartbeat."
plot_attention_heatmap(text3)
```
